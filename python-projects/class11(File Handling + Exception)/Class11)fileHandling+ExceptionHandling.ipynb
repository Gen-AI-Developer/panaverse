# Pass by reference vs value in Python

https://www.geeksforgeeks.org/pass-by-reference-vs-value-in-python/

```python
def call_by_value(x):
    x = x * 2
    print("in function value updated to", x)
    return
 
def call_by_reference(list):
    list.append("D")
    print("in function list updated to", list)
    return
 
my_list = ["E"]
num = 6
print("number before=", num)
call_by_value(num)
print("after function num value=", num)
print("list before",my_list)
call_by_reference(my_list)
print("after function list is ",my_list)
```

# Error Handling

try:
    body
except:
    body
else:
    body
finally:
    body

```python
l1 : list[int] = [1,2,3,4,5,6]

try:
    print(f'Value at Index Number : { l1[int(input("Please enter index"))]}')
except:
    print(f'Your Code Rain into a problem')
```

# Best way to Handle Exception in Python

```python
l1 : list[int] = [1]

try:
    print(f'Value at Index Number : { l1[int(input("Please enter index"))]}')
except Exception as e:
    print(f'Your Code fall into a problem :- {e}')
```
